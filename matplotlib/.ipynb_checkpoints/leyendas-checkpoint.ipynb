```python
PotashAvg = pot['2012':].Close.mean(axis=1)
PotashStd = pot['2012':].Close.std(axis=1)
Potash = pd.DataFrame(index = pot['2012':].index)
Potash['price']= pot['2012':].Close
Potash['average']= pot['2012':].Close.mean(axis=1)
Potash['SD1']= pot['2012':].Close.std(axis=1)+pot['2012':].Close.mean(axis=1)
Potash['mSD1']= -pot['2012':].Close.std(axis=1)+pot['2012':].Close.mean(axis=1)
Potash['SD2']= 2*pot['2012':].Close.std(axis=1)+pot['2012':].Close.mean(axis=1)
Potash['mSD2']= -2*pot['2012':].Close.std(axis=1)+pot['2012':].Close.mean(axis=1)

mpl.rcParams['font.family']='Arial'
mpl.rcParams['font.size']=9
fig = plt.figure()
ax = plt.subplot(111)

ax.plot(Potash.index, Potash.price, color='red', linewidth=2)
ax.plot(Potash.index, Potash.average, color='black', linewidth=1)
ax.plot(Potash.index, Potash.SD1, '-.', color = 'black',linewidth=0.9 )
ax.plot(Potash.index, Potash.SD2, ':', color = 'black',linewidth=0.9 )
ax.plot(Potash.index, Potash.mSD1, '--', color = 'black',linewidth=0.9 )
ax.plot(Potash.index, Potash.mSD2, '-.', color = 'black',linewidth=0.9 )

Potash.columns = ['POT', 'Average','+1 SD', '-1 SD', '+2 SD', '-2 SD']
lg = ax.legend(Potash.columns, bbox_to_anchor=(0.5,-0.05),loc='upper center', ncol=6, columnspacing=0.1, handletextpad=0)
lg.draw_frame(False)
myFmt = mdates.DateFormatter('%b-%y')
ax.xaxis.set_major_formatter(myFmt)

# Set the fontsize
for label in lg.get_texts():
    label.set_fontsize(9)

for label in lg.get_lines():
    label.set_linewidth(1.5)  # the legend line width
    

plt.savefig('pot')
plt.show()
```
