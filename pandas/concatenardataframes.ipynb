```python
# Genero fechas hasta una fecha, antes que hoy, esta serie ya hice con dias laborales
dates = pd.date_range('1/1/2004','15/10/2010',freq='B') # B es business day
```

```python
dates
```

```python
#datos aleatorios para ambas series
data = [np.random.randint(20, 100) for x in range(len(dates))]
data2 = [np.random.randint(20, 100) for x in range(len(dates))]
```

```python
# genero dos series para hacer el dataframe df, seguro hay una forma más inteligente de hacer esto, pero funciona

s1 = pd.Series(data, index=dates)
s2 = pd.Series(data2, index=dates)
df = pd.concat([s1, s2], axis=1)
df.columns = ['azul','rojo']
df.head()
```

```python
#Con BDay(1) agrego un dia laboral, hay que agregar from pandas.tseries.offsets import BDay
#sino la cosa no anda
NuevaFecha = df.index.max()+BDay(1)
Hoy = datetime.datetime.now()
NuevaFecha, Hoy
```

```python
#ahora me genero un nuevo dataframe desde la fecha nueva, es decir un día laboral más que el ultimo del dataframe
#original, hasta hoy... da lo mismo cuando... y si se, con Control m m puedo por comentarios mas fancy
#en un notebook, pero me da lata

dates3 = pd.date_range(NuevaFecha,Hoy,freq='B') # B es business day
data3 = [np.random.randint(20, 100) for x in range(len(dates3))]
data4 = [np.random.randint(20, 100) for x in range(len(dates3))]
s3 = pd.Series(data3, index=dates3)
s4 = pd.Series(data4, index=dates3)
df2 = pd.concat([s3, s4], axis=1)
df2.columns = ['azul','rojo']
```

```python
df2.head(), df2.tail() #esto es para ver que este bien la cosa
```

```python
#y ahora fusiono ambos dataframe y estamos listocos, con el axis =0 se enchufa abajo, con axis = 1 queda la cagada
#porque se pone al lado
mergedpd=pd.concat([df, df2], axis=0)
```
