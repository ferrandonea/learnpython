### Solo para huevear voy a agregar movimiento browniano a esta cosa

lo mismo de arriba, pero con el browniano dichoso

```python
def geometric_brownian_motion(T = 1, N = 100, mu = 0.1, sigma = 0.01, S0 = 20):
    "http://stackoverflow.com/a/13203189/190597 (unutbu)"
    dt = float(T)/N
    t = np.linspace(0, T, N)
    W = np.random.standard_normal(size = N) 
    W = np.cumsum(W)*np.sqrt(dt) ### standard brownian motion ###
    X = (mu-0.5*sigma**2)*t + sigma*W 
    S = S0*np.exp(X) ### geometric brownian motion ###
    return S

dates = pd.date_range('2012-01-01', '2013-02-22')
T = (dates.max()-dates.min()).days / 365
N = dates.size
start_price = 100
y = pd.Series(
    geometric_brownian_motion(T, N, sigma=0.1, S0=start_price), index=dates)
y.plot(color='red')

start_price = 110
y2 = pd.Series(
    geometric_brownian_motion(T, N, sigma=0.1, S0=start_price), index=dates)
y2.plot(color='black')

plt.show()
```

```python
NuevaFecha = y.index.max()+BDay(1)
Hoy = datetime.datetime.now()
NuevaFecha, Hoy
```

```python
dates = pd.date_range(NuevaFecha, Hoy)
T = (dates.max()-dates.min()).days / 365
N = dates.size
start_price = y.iloc[-1]
y3 = pd.Series(
    geometric_brownian_motion(T, N, sigma=0.1, S0=start_price), index=dates)
y3.plot(color='red', linestyle='dashed')

start_price = y2.iloc[-1]
y4 = pd.Series(
    geometric_brownian_motion(T, N, sigma=0.1, S0=start_price), index=dates)
y4.plot(color='black', linestyle='dashed')
```
